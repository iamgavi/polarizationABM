# robustness - dimensions

```python
path = './plot_raw_data/robustness/'

fig, axes = plt.subplots(1, 2, figsize=(10, 5))


for i, name in zip(range(2)[::-1], ['robustness_dim_alpha.tsv', 'robustness_dim_beta.tsv']):
    f = pd.read_csv(path + name, delimiter=',', header=0)
    for width, dim in list(zip([1.5, 1.5, 2, 3, 5], [1, 2, 4, 10, 20]))[::-1]:
        dt_sub = f[f['dim'] == str(dim)]
        if i == 1:
            axes[i].plot([float(s) for s in dt_sub.x], 
                         [float(s) for s in dt_sub.y], label="{} issues".format(dim), 
                        linewidth=width)
            axes[i].set_xlabel('Party Identity ({})'.format(chr(946)))
            axes[i].set_ylabel('Intolerance ({})'.format(chr(945)))
        else:            
            axes[i].plot([float(s) for s in dt_sub.y], 
                         [float(s) for s in dt_sub.x], label="{} issues".format(dim), 
                        linewidth=width)
            axes[i].set_ylabel('Party Identity ({})'.format(chr(946)))
            axes[i].set_xlabel('Intolerance ({})'.format(chr(945)))
        
    #axes_prop(axes[i])
    axes[i].legend()
    if i == 1:
        axes[i].set_xticklabels([0, .2, .4, .6, .8, 1])
        axes[i].set_xticks([0, 2, 4, 6, 8, 10])
        axes[i].set_yticklabels([0, .2, .4, .6, .8])
        axes[i].set_yticks([0, 20, 40, 60, 80])
        axes[i].text(-3, 80, 'B', fontsize=18, fontweight=1000, fontfamily='opensans-extrabold')
    else:
        axes[i].set_yticklabels([0, .2, .4, .6, .8])
        axes[i].set_yticks([0, 20, 40, 60, 80])
        axes[i].set_xticklabels([0, .2, .4, .6, .8, 1])
        axes[i].set_xticks([1, 3, 5, 7, 9, 11])
        axes[i].text(-2, 80, 'A', fontsize=18, fontweight=1000, fontfamily='opensans-extrabold')

        
fig.tight_layout()  # need this one to get rid of the paddings
fig.savefig('plot_raw_data/robustness_dim.png', dpi=300, bb_inches='tight')
        
```

# robustness - 30r

```python
path = './plot_raw_data/robustness/'

fig, axes = plt.subplots(1, 2, figsize=(10, 4))


for i, name in zip(range(2), ['robustness_30r_alpha.tsv', 'robustness_30r_beta.tsv']):
    f = pd.read_csv(path + name, delimiter=',', header=0)
    
    for color, direction, width, linestyle in list(zip(['r', 'b'], 
                                   ['->', '<-'], 
                                   [.01, .01], 
                                 ['-', '.-'])):
        dt_sub = f[(f['direction'] == direction)]
        dt_sub['x'] = np.array(dt_sub.loc[:, ['x']]).astype(float)
        dt_sub['y'] = np.array(dt_sub.loc[:, ['y']]).astype(float)
        if i == 0:
            dt_sub = dt_sub[(abs(dt_sub['x']/0.05 - dt_sub['x']//0.05) < 0.1) | 
                           (abs(dt_sub['x']/0.05 - dt_sub['x']//0.05) > 0.9)]
            dt_sub = dt_sub.groupby("x").sample(n=25, random_state=1).reset_index()
            
            
        axes[i].scatter(dt_sub.x, 
                         dt_sub.y, 
#                     label="m={}".format(log_s), 
                    color=color, linewidth=width, marker='.')
        dt_sub = dt_sub.groupby('x')['y'].mean().reset_index()
        axes[i].plot(dt_sub.x, 
                 dt_sub.y, 
#                     label="m={}".format(log_s), 
            color=color, linewidth=1)

    
    axes[i].set_xticks([0, .2, .4, .6, .8, 1])
    axes[i].set_xticklabels(['0', '.2', '.4', '.6', '.8', '1'])
    axes[i].set_yticks([0, .2, .4, .6, .8, 1])
    axes[i].set_yticklabels(['0', '.2', '.4', '.6', '.8', '1'])
    
    
        
    #axes_prop(axes[i])
    
    if i == 0:
        axes[i].set_xlabel('Intolerance ({})'.format(chr(945)))
        axes[i].text(.9, .85, "{}=0.5".format(chr(946)), fontsize=14)
        axes[i].text(-0.23, 0.98, 'A', fontsize=14, fontweight=1000, fontfamily='opensans-extrabold')
        axes[i].arrow(0.35, 0.3, 0, 0.4, 
                  head_width = 0.02,
                  width = 0.002,
                  ec ='r', 
                  fc='r')
        axes[i].text(.4, .35, "polarization", rotation=90, color='r')
    else:
        axes[i].set_xlabel('Party Identity ({})'.format(chr(946)))
        axes[i].text(.9, .85, "{}=0.5".format(chr(945)), fontsize=14)
        axes[i].text(-0.23, 0.98, 'B', fontsize=14, fontweight=1000, fontfamily='opensans-extrabold')
        axes[i].arrow(0.4, 0.3, 0, 0.4, 
                  head_width = 0.02,
                  width = 0.002,
                  ec ='r', 
                  fc='r')
        axes[i].text(.45, .35, "polarization", rotation=90, color='r')
    axes[i].set_ylabel('Extremism')

fig.tight_layout()  # need this one to get rid of the paddings    
fig.savefig('plot_raw_data/robustness_20r.png', dpi=300, bb_inches='tight')
```
